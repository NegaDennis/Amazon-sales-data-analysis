# Data Cleaning

## Data format

```python
# Checking format of ratings
df['ratings'].unique()
```

```python
# Dropping row with incorrect format
to_drop = ['FREE','₹68.99','₹65','₹70','₹100','₹99','₹2.99','Get']
df = df[~df['ratings'].isin(to_drop)]
print(df['ratings'].unique())
print(df.shape)
print(df['ratings'].dtype)
```

```python
# Converting all to numeric, forcing errors to NaN
df['ratings'] = pd.to_numeric(df['ratings'], errors='coerce')

# Converting to integer type that allows NaNs
df['ratings'] = df['ratings'].astype('float64')
```

```python
# Checking format of no_of_ratings
df.sort_values(by='discount_price',ascending=True)
```

```python
# Cleaning up price-related columns
prices = ['discount_price','actual_price']
for col in prices:
  df[col] = df[col].str.replace('₹', '', regex=False)   #Removing unit of currency
  df[col] = df[col].str.replace(',', '', regex=False)   #Removing commas (because they were randomly added)
  df[col] = df[col].astype('float64')
  df = df.rename(columns={col:'{} (₹)'.format(col)})               # Adding currency unit to headers

df.head(10)
```

```python
# Checking if there's inconsistent in data type in no_of_ratings
df['no_of_ratings'].apply(lambda x: not isinstance(x, (float, int))).any()
```

```python
# Filter out errors
df[df['no_of_ratings'].apply(lambda x: isinstance(x, (float, int)))]
```

By accessing via the included links, it seem like most of these products only have very fews to no ratings at all. It can be theorized that at the time of data scraping, they had no ratings at all. It explains behaviours of both ratings and no_of_ratings. I will look for exceptions next.

```python
# Inspecting cases where only ratings or no_of_ratings is empty
df[(df['ratings'].isnull()==True) & (df['no_of_ratings'].isnull() == False)].sort_values(by='name')
```

```python
# Inspecting cases where only ratings or no_of_ratings is empty
df[(df['ratings'].isnull()==False) & (df['no_of_ratings'].isnull() == True)].sort_values(by='name')
```

No exceptions were detected. We will impute missing no_of_ratings with 0 to reflect our theory and to make conversion possible.

```python
# Converting all to numeric, forcing errors to NaN
df['no_of_ratings'] = pd.to_numeric(df['no_of_ratings'], errors='coerce')

# Filling real NaNs with 0 as said before
df['no_of_ratings'] = df['no_of_ratings'].fillna(0)

# Converting to integer type that allows NaNs
df['no_of_ratings'] = df['no_of_ratings'].astype('int64')
```
